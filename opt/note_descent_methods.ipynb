```julia
using Pkg; Pkg.activate(".")
using Revise
import Plots: @animate, plot3d, gif, @gif
import PyPlot
import PyPlot: subplot, title, xlabel, ylabel, ylim, xlim, plot, savefig
import LinearAlgebra: norm, dot
import Base
import SparseArrays: spdiagm, sprand, spzeros

include("./src/Opt.jl")
import .Opt
```

```julia
n = Int64(1e2)
f, grad! = Opt.hard_leastsquares_problem(n)

x = sprand(n, 0.5)
g = similar(x)
grad!(g,x)

@show f(x)
@show size(g)
```

```julia
x0 = spzeros(n)
n_iterations = 1000

Γf = zeros(n_iterations, n+1)
grad_norm = zeros(n_iterations)

function access_state(state)
    Γf[state.k,:] = [state.x..., state.f]
    grad_norm[state.k] = norm(state.g)
end

Opt.gradient_descent(x0, f, grad!;
    α = 0.001,
    n_iterations = n_iterations,
    access_state = access_state)
```

```julia
subplot(211)
PyPlot.plot(1:n_iterations, Γf[:,end])
title("f(x)")
xlabel("k")

subplot(212)
PyPlot.plot(1:n_iterations, grad_norm)
title("|∇f(x)|")
xlabel("k")
```

```julia
x0 = spzeros(n)
n_iterations = 1000
α₀ = 0.01
barzilaiborwein_type = 1
g_abstol = 1e-8
access_state(state) = nothing

x = copy(x0)
g = copy(x0); grad!(g, x)
u, v = zeros(size(x0)), zeros(size(x0))

s = Opt.GradientDescentBarzilaiBorweinState(x0)
grad!(s.g, x)

for k in 1:n_iterations
    
    s.k, s.f = k, f(x)
    @. s.x_prev = s.x
    @. s.x = x
    @. s.g_prev = s.g
    @. s.g = g

    if k == 1
        # should do line search
        s.α = α₀
    else
        u = s.x - s.x_prev
        v = s.g - s.g_prev
        s.α = (barzilaiborwein_type == 1) ?
            dot(u, v) / norm(v)^2 :
            norm(u)^2 / dot(u, v)
    end

    access_state(s)        

    x = s.x + s.α*(-s.g)
    grad!(g, x)

    if norm(s.g) <= g_abstol
        println("Terminate at k=$k: |∇f(x)| = $(norm(s.g)) <= $g_abstol")
        break
    end
end
```

```julia
x0 = spzeros(n)
n_iterations = 10
fs = zeros(n_iterations)
grad_norm = zeros(n_iterations)
αs = zeros(n_iterations)

function access_state(state)
    fs[state.k] = state.f
    grad_norm[state.k] = norm(state.g)
    αs[state.k] = state.α
end

Opt.gradient_descent_barzilaiborwein(x0, f, grad!;
    n_iterations = n_iterations,
    access_state = access_state)
```

```julia
N = 10
xs = 2:N

subplot(221)
PyPlot.plot(xs, fs[xs])
ylabel("f(x)")
xlabel("k")

subplot(222)
PyPlot.plot(xs, grad_norm[xs])
ylabel("|∇f(x)|")
xlabel("k")

subplot(223)
PyPlot.plot(xs, αs[xs])
ylabel("α")
xlabel("k")
```

```julia
# show asymptotic equivalence of 
#     β1_k = (t_{k-1}] - 1)/t_k for t_1 = 1, and t_{k+1} = (1 + √(1 + 4t_k^2))/2
#     β2_k = k/(k+3)

n_iterations = 200

ts = zeros(n_iterations)
β1 = zeros(n_iterations)
β2 = zeros(n_iterations)
β3 = zeros(n_iterations)
β4 = zeros(n_iterations)

for k in 1:n_iterations
    if k == 1
        t_prev = 1
    else
        t_prev = ts[k-1]
    end
    
    ts[k] = 0.5(1 + √(1 + 4t_prev^2))
    
    β1[k] = (t_prev - 1)/ts[k]
    β2[k] = k / (k+3)
    β3[k] = (k-1)/(k+2)
    β4[k] = (k-3)/k
end

xs = 1:n_iterations 
fig, ax = PyPlot.subplots()
ax.spines["right"].set_visible(false)
ax.spines["top"].set_visible(false)
PyPlot.plot(xs, β1, label="(t_k-1)/t_{k+1}")
PyPlot.plot(xs, β2, label="k/(k+3)")
PyPlot.plot(xs, β3, label="(k-1)/(k+2)")
PyPlot.plot(xs, β4, label="(k-3)/k")
title("Asymptotic behavior of momentum term βₖ")
ylabel("βₖ")
xlabel("k")
ylim(0,1)
xlim(0,50)
PyPlot.legend()
savefig("summary/assets/asymptotic_behavior_of_momentum_term.png")
```

```julia
x0 = spzeros(n)
n_iterations = 1000
α = 0.001
g_abstol = 1e-8
access_state(state) = nothing

x = copy(x0)
y = copy(x0)
s = Opt.GradientDescentNesterovState(x0)

for k in 1:n_iterations

    s.k, s.f = k, f(x)
    s.x .= x
    s.y .= y
    grad!(s.g, y)

    access_state(s)

    @. x = s.y + α*(-s.g)
    @. y = x + (k-1)/(k+2)*(x - s.x)

    if norm(s.g) <= g_abstol
        println("Terminate at k=$k: |∇f(yᵏ)| = $(norm(s.g)) <= $g_abstol")
        break
    end
end
```
